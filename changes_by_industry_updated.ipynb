```python
aus_df.head()
```

```python
aus_df_ft = aus_df.loc[:,[c for c in aus_df.columns if "full-time" in c]]
aus_df_pt = aus_df.loc[:,[c for c in aus_df.columns if "part-time" in c]]
```

```python
aus_df_ft = aus_df_ft.iloc[:,:-1]
aus_df_pt = aus_df_pt.iloc[:,:-1]
aus_df_ft.head(2)
```

```python
# rename columns to select only the field name
aus_df_ft.columns = [c.split(";")[1] for c in aus_df_ft.columns]
aus_df_pt.columns = [c.split(";")[1] for c in aus_df_pt.columns]
```

```python
# find 5 increasing field and 5 decreasing field : before covid vs after covid

aus_df_ft_18 = aus_df_ft.loc["2018-01-01":"2019-12-31"]
aus_df_ft_19 = aus_df_ft.loc["2020-01-01":"2022-12-31"]


op = aus_df_ft_19.mean(axis=0) - aus_df_ft_18.mean(axis=0)
```

```python
op.sort_values()
```

```python
# select top 5 and bottom 5
top_ind = list(op.sort_values(ascending=False).iloc[:5].index)
bottom_ind = list(op.sort_values().iloc[:5].index)

op = op.loc[top_ind + bottom_ind]
op
```

```python
# Plot the top 5 and bottom 5
plt.figure(figsize=(16,8))
plt.bar(op.index[:], op.values[:])
plt.xticks(rotation=90);
plt.title("Australia Full time employement: Before covid vs After covid: Top increasing field and decreasing field")

```
